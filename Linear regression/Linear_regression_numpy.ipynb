```python
# import libraries 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# initialize X and Y 
X = []
Y= []

for line in open('data_1d.csv'):
  x,y = line.split(",")
  X.append(float(x))
  Y.append(float(y))
```

```python
# Lets turn the arrays into numpy arrays instead 

X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X,Y)
```

```python
denominator = X.dot(X) - (X.mean() * X.sum())
```

```python
a = (X.dot(Y)-X.mean()*Y.sum())/ denominator
b = ((Y.mean()*X.dot(X)) - (X.mean()*X.dot(Y))) / denominator
```

```python
Yhat = a*X + b
```

```python
plt.scatter(X,Y)
plt.plot(X, Yhat)
plt.show
```

## Determine how good the model is using R squared 

```python
# Calculate r-squared 

d1  =Y - Yhat
d2 = Y - Y.mean()

r2 = 1 - d1.dot(d1) / d2.dot(d2)
```
